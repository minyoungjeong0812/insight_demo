```python
df1['kill_per_min']=(df1['kill']/df1['playtime'])*60
```

```python
df2
```

```python
df3
```

```python
kpm_1=df1["kill_per_min"]
```

```python
kpm_2=df2["kill_per_min"]
kpm_3=df3["kill_per_min"]
```

```python
kpm_1
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(font_scale = 1.7)
sns.distplot(kpm_1, label='regular')
sns.distplot(kpm_2, label='cheater')
sns.distplot(kpm_3, label = 'pro')
plt.legend()
```

```python
df1['head_per_kill'] = df1['headshot']/df1['kill']
df2['head_per_kill'] = df2['headshot']/df2['kill']
df3['head_per_kill'] = df3['headshot']/df3['kill']
```

```python
df1['head_per_kill']=df1['head_per_kill'].fillna(0)
```

```python
df2['head_per_kill']=df2['head_per_kill'].fillna(0)
df3['head_per_kill']=df3['head_per_kill'].fillna(0)
```

```python
hpk_1=df1["head_per_kill"]
hpk_2=df2["head_per_kill"]
hpk_3=df3["head_per_kill"]
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(font_scale = 1.7)
sns.distplot(hpk_1, kde='False',label='regular')
sns.distplot(hpk_2, kde='False',label='cheater')
sns.distplot(hpk_3, label = 'pro',kde='False')
plt.legend()
```
